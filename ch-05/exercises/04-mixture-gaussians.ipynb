```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0.8, 0.5, size=(1024,))
y = np.random.normal(-0.8, 0.5, size=(1024,))

plt.hist(x, bins=100)
plt.hist(y, bins=100)
plt.show()
```

```python
lambda_ = 0.6
xy = lambda_ * x + (1 - lambda_) * y
xy = np.unique(xy)
ln_xy = -np.log(xy)

plt.plot(xy, ln_xy)
plt.title('-ln(x) is convex')
plt.show()
```

```python
import math

lambd = 0.5

print(math.log(0.5*math.exp(0.5**2) + 0.5*math.exp(0.5**2)))
print(0.5*math.log(math.exp(0.5**2)) + 0.5*math.log(math.exp(0.5**2)))
```

```python
x = np.linspace(0.1, 1.0, num=100)
y = np.linspace(0.5, 1.5, num=100)

lambd = 0.9
ex = np.exp(x**2)
ey = np.exp(y**2)
log_exey = np.log(lambd * ex + (1 - lambd) * ey)
ex_log = lambd * np.log(ex) + (1 - lambd) * np.log(ey)


plt.plot(log_exey, ex_log)
plt.show()
```
