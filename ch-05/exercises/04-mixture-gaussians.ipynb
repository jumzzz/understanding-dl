```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0.8, 0.5, size=(1024,))
y = np.random.normal(-0.8, 0.5, size=(1024,))

plt.hist(x, bins=100)
plt.hist(y, bins=100)
plt.show()
```

```python
lambda_ = 0.6
xy = lambda_ * x + (1 - lambda_) * y
xy = np.unique(xy)
ln_xy = np.log(xy)

plt.plot(xy, ln_xy)
plt.title('ln(x) is concave')
plt.show()
```
