### On Optimizing on the Mean

We have encountered predictions like these multiple times already

```python
import numpy as np
import math
import matplotlib.pyplot as plt

X = np.linspace(-2 * math.pi, 2 * math.pi)
y_target = np.cos(X)
y_pred = np.mean(y_target)

plt.plot(X, y_target, color='blue',label='Target')
plt.axhline(y_pred, color='red', label='Prediction')
plt.title('Figure 1: Target VS Learned Prediction')
plt.legend()
plt.show()
```

### Explaining Figure 1
- The reason why this happens is simply because MSE Loss function is derived by minimizing the predicted Mean Squared Error. Hence, the optimizer will seek to predict the mean of the data as a shortcut.

### Interesting Notes to Reflect Upon
> To make two or more prediction types simultaneously, we similarly assume the errors in each are independent. For example, to predict wind direction and strength, we might choose the von Mises distribution (defined on circular domains) for the direction and the exponential distribution (defined on positive real numbers) for the strength. The independence assumption implies that the joint likelihood of the two predictions is the product of individual likelihoods. These terms will become additive when we compute the negative log-likelihood.

### Reformulating negative log-loss as Cross Entropy Loss
> It follows that the negative log-likelihood criterion (from maximizing the data likelihood) and the cross-entropy criterion (from minimizing the distance between the model
and empirical data distributions) are equivalent.
